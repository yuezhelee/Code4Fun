```python
import tweepy
import pandas as pd


# the following info is created to pull data from Trump's Twitter
consumer_key= 'yOURkEY'
consumer_secret= 'YOURkEY'
access_token= 'yourkEY'
access_token_secret= 'yoURKEy'
```

```python
def get_all_tweets(screen_name):

    #authorize twitter, initialize tweepy
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth)

    #initialize a list to hold all the tweepy Tweets
    alltweets = []

    #make initial request for most recent tweets (200 is the maximum allowed count)
    new_tweets = api.user_timeline(screen_name = screen_name,count=200)
    
    #save most recent tweets
    alltweets.extend(new_tweets)
    
    #save the id of the oldest tweet less one
    oldest = alltweets[-1].id - 1

    #keep grabbing tweets until there are no tweets left to grab
    while len(new_tweets) > 0:
        new_tweets = api.user_timeline(screen_name = screen_name,count=200,max_id=oldest)
        alltweets.extend(new_tweets)
        oldest = alltweets[-1].id - 1

        outtweets = [tweet.text.encode("utf-8") for tweet in alltweets]
        
    return outtweets
```

```python
trumptweet = get_all_tweets("realDonaldTrump")
```

```python
cleanedtweet = [str(w).replace("b'", "") for w in trumptweet]
cleanedtweet
```

```python
# analysis of word frequencies
import nltk
from nltk.stem import WordNetLemmatizer

tweettext = ''.join(map(str, cleanedtweet))

text_tmp = nltk.word_tokenize(tweettext)
dist = nltk.FreqDist(text_tmp)
dist['Russia'] + dist['Russian']
```
